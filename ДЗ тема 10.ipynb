```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
#    'events.json'
with open(file='events.json') as f:
    events = json.loads(f.read())

#   DataFrame
df = pd.DataFrame(events["events"])

#
plt.figure(figsize=(10, 5)) # ""    10x5
sb.countplot(data = df, x = "signature") #     ,    df    x  "signature"

plt.title("    ") #  
plt.xticks(rotation = 30, ha = 'right') #   x   30  
plt.show()
```
