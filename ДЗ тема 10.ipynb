```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
#Загрузка данных из файла 'events.json'
with open(file='events.json') as f:
    events = json.loads(f.read())

#Преобразование данных в DataFrame
df = pd.DataFrame(events["events"])

#Визуализация
plt.figure(figsize=(10, 5)) #Создание "холста" для графика размером 10x5
sb.countplot(data = df, x = "signature") #Использование функции для построения графика счетчика, заполнение его данными df и задание оси x по "signature"

plt.title("Распределения типов событий информационной безопасности") #Добавление названия графика
plt.xticks(rotation = 30, ha = 'right') #Разворот подписей у x оси на 30 градусов направо
plt.show()
```
